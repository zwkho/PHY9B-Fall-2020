```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun(x):
    if x < -1:
        return 0
    elif -1 <= x < 0:
        return 1+x
    elif 0 <= x < 1:
        return 1-x
    elif x > 1:
        return 0
```

```python
x_values = np.linspace(-4,4, 1000)
```

```python
plt.plot(x_values, [fun(x) for x in x_values])
```

```python
plt.plot(x_values, [fun(x-1) for x in x_values])
```

```python
plt.plot(x_values, [fun(x) for x in x_values])
plt.plot(x_values, [fun(x-1) for x in x_values])
plt.plot(x_values, [fun(x-2) for x in x_values])
```

```python
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
fig, ax = plt.subplots()
ax.set_xlim((-1, 8))
ax.set_ylim((0, 1))

line, = ax.plot([], [], lw=2)

def init():
    line.set_data([], []);
    return (line,)

def animate(i):
    v = 0.01
    x = np.linspace(-1, 8, 1000)
    y = [fun(xx - v*i) for xx in x]
    line.set_data(x, y)
    return (line,)

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=500, interval=20, blit=True, repeat = False)
```

```python
HTML(anim.to_html5_video())
```

```python
xs = np.linspace(-10,10, 100)
y = np.sin((2*np.pi/10)*xs)
y2 = -np.sin((2*np.pi/10 )*xs + np.pi/2)
plt.figure(figsize=(15,3))
for i in xs:
    plt.axvline(i, ymin = 0.4, ymax = 0.6, color = 'black');
plt.axis('off');

displaced = xs + y
plt.figure(figsize=(15,3))
for i in displaced:
    plt.axvline(i, ymin = 0.4, ymax = 0.6, color = 'black');
plt.axis('off');
plt.plot(xs, y, label = 'Displacement')
plt.plot(xs, y2, label = 'Pressure')
plt.legend()
```
